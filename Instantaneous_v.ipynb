## Building a graph of change in distance over time and instantaneous velocity using Calculus

```python
# importing the relevant libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# building the function that defines the relationship between distance and time (the variable factor)
def q(x):
  return x**3 + 4*x + 2
```

```python

t = [1,2,3,4,5,6,7] # I made a list of different point in time in seconds
d = [] # for the distance covered at the different point in time
# a for loop that calculates the distance at each point in time
for i in t:
  c = q(i)
  d.append(c)
print(d)
```

```python
# a plot of the relationship between distance and time
plt.plot(t, d)
plt.xlabel('time')
plt.ylabel('distance')
plt.show()
```

```python
# finding the instantaneous velocities
iv = []
for i in range(len(t)):
  new_t = t[i] + 0.0001
  new_d = q(new_t)
  dd_dt = (new_d - d[i] ) / (new_t - t[i])
  iv.append(dd_dt)
iv
```

```python
# a plot of realtionship between time and distance, also the relationship between time and instantaneous velocity
plt.plot(t,d, label='distance/time')
plt.plot(t, iv, label='instantaneous velocity over time')
plt.xlabel('time')
plt.ylabel('distance and velocity')
plt.legend()
plt.show()
```
